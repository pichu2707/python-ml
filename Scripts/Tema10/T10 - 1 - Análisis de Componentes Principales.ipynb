### 1- Calculamos la descomposición de valores y vectores propios

#### a) Usando matriz de covarianza

```python
from IPython.display import display, Math, Latex
```

```python
display(Math(r'\sigma_{jk} = \frac{1}{n-1}\sum_{i=i}^m (x_{ij}) - \overline{x_j})(x_ik)-\overline{x_k}'))
```

```python
display(Math(r'\Sigma = \frac{1}{n-1}((X-\overline{x})^T(X-\overline{x}))'))
```

```python
display(Math(r'\overline{x} = \sum_{i=1}^n x_i\in \mathbb R^m'))
```

```python
import numpy as np
```

```python
mean_vect = np.mean(X_std, axis=0)
mean_vect
```

```python
cov_matrix = (X_std - mean_vect).T.dot((X_std - mean_vect))/(X_std.shape[0]-1)#Matrixz de covarianzas creado por nosotros
print(" La matriz de covariancias es:  \n%s"%cov_matrix)
```

```python
np.cov(X_std.T)#Matriz de covarianzas de la que partimos con numpy
```

```python
eig_vals, eig_vectors = np.linalg.eig(cov_matrix)
print("Valores propios: \n%s"%eig_vals)
print("VAlores propios: \n%s "%eig_vectors)
```

#### b) Usando la matriz de correlación
Es más usado en el campo de las finanzas

```python
corr_matrix = np.corrcoef(X_std.T)
corr_matrix
```

```python
eig_vals_corr, eig_vectors_corr = np.linalg.eig(corr_matrix)
print("Valores propios: \n%s"%eig_vals_corr)
print("VAlores propios: \n%s "%eig_vectors_corr)
```

#### c) Singular Value Decomposition

```python
u,s,v = np.linalg.svd(X_std.T)
u
```

### 2. - Las componentes principales

```python
for ev in eig_vectors:
    print("La longitud del vector propio es: %s"%np.linalg.norm(ev))
```

```python
eigen_pairs = [(np.abs(eig_vals[i]),eig_vectors[:,i]) for i in range(len(eig_vals))]
eigen_pairs
```

```python
#Ordenamos los vectores propios con valores propios de mayor a menor
```

```python
eigen_pairs.sort()
eigen_pairs.reverse()
eigen_pairs
```

```python
print("Valores propios en orden descendente: ")
for ep in eigen_pairs:
    print(ep[0])
```

```python
total_sum = sum(eig_vals)
var_exp = [(i/total_sum)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
var_exp
```

```python
plot1 = Bar(x=["CP %s"%i for i in range(1,5)], y = var_exp, showlegend=False)
plot2 = Scatter(x=["CP %s"%i for i in range(1,5)], y = cum_var_exp, showlegend=True, name = "porcentaje de varianza acumulada")
data = Data([plot1, plot2])
layout = Layout(xaxis = XAxis(title="Componentes principales"),
               yaxis = YAxis(title = "Porcentaje de varianza aplicada"),
               title= "Porcentaje de variabilidad explicada por cada componente principal")

fig=go.Figure(data=data, layout=layout)
chart_studio.plotly.iplot(fig)
fig.show()
```

```python
#Vamos a reducri un espacio vectorial de 4 a 2 vectores propios escogiendo lo s que más nos interesan
W = np.hstack((eigen_pairs[0][1].reshape(4,1),
              eigen_pairs[1][1].reshape(4,1)))
W
```

```python
display(Math(r'Y = X \cdot W, X \in M(\mathbb R)_{150, 4}, W \in M(\mathbb R)_{4,2}, Y \in M(\mathbb R)_{150, 2}'))
```

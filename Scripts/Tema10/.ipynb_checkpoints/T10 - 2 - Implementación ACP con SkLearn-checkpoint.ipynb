```python
import pandas as pd
import chart_studio
import plotly.offline as plt
from plotly.graph_objs import *
import plotly.tools as tls
import plotly.graph_objects as go

from sklearn.preprocessing import StandardScaler

chart_studio.tools.set_credentials_file(username='pichu_2707', api_key='Jl2zwifKC8BC3ikpSJl2')
```

```python
df = pd.read_csv("C:/Users/Megaport/Desktop/python/datos machine learning/iris/iris.csv")
```

```python
X = df.iloc[:,0:4].values
y = df.iloc[:,4].values
X_std = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA as sk_pca
```

```python
acp = sk_pca(n_components=2)#cogemos dos porque en el T10 - 1 - ya hemos visto que solo nos hacen falta dos modelos
Y = acp.fit_transform(X_std)
Y
```

```python
results = []
for name in ('setosa','versicolor','virginica'):
    result = Scatter(x = Y[y == name, 0], y = Y[y==name, 1],
                   mode = "markers", name = name,
                   marker = Marker(size=15, line=Line(color="rgba(225,225,225,0.2)", width=0.5),
                                  opacity=0.75)) 
    results.append(result)

data=Data(results)
layout = Layout(xaxis=XAxis(title="CP1", showline=False),
               yaxis=YAxis(title="CP2", showline=False))

fig=go.Figure(data=data, layout=layout)
chart_studio.plotly.iplot(fig)
fig
```

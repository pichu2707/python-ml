```python
#Función que genera una lista de n números aleatorios entero dentro del intervalo [a,b]
def randint_list (n, a, b):
    x=[]
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(25, 1, 50)
```

## Simulación de Monte carlo

* Generamos dos números aleatorios uniforme x e y entre 0 y 1
* Calcularemos x*x+y*y
    * Si el valor es inferior a 1 -> estoy dentro del círculo
    * Si el valor es superior a 1 -> estoy fuera del círculo
* Calculamos el número total de veces que están dentro del círculo y lo dividimos entre el número total de intentos para obtener una aproximación de la probabilidad de caer dentro del círculo
* Usamos dicha probabiliidad para aproximar el valor de pi.
* Repetimos el experimento número suficiente de veces (por ejemplo 1000) para obtener (1000) diferentes aproxiamaciónes a pi.
* Calculamos el promedio de los 1000 experimentos para dar un valor final de pi.

```python
def pi_montecarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []
    for i in range(100):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt(x[j] * x[j] + y[j] * y[j])
            if z<=1:
                value += 1
        float_value = float(value)
        pi_value=float_value * 4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value

    pi = pi_avg/100

    print(pi)

    fig = plt.plot(pi_value_list)
    return(pi, fig)
```

```python
pi_montecarlo(100000, 2000)
```

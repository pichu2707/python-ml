# Lectura de datos

```python
import numpy as np
import pandas as pd
```

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
import csv
#importamos la librería
import urllib3
```

```python
#Hacemos la conexión con la web de los datos
http = urllib3.PoolManager()
r = http.request('GET', medals_url)
r.status
response = r.data
```

```python
cr = csv.reader(response)
```

```python
cr
```

```python
response
```

```python
#El objetivo reponse contiene un string binartio, así que lo convertimos a un string decodificándolo a UTF-8
str_data = response.decode('utf-8')
```

```python
#Dividimos el string en un array de filas, separándolo por los intros "\n"
lines = str_data.split("\n")
lines
```

```python
#La primera linea contiene la cabecera, así que hay que extraerla
col_names = lines[0].split(",")
n_cols = len(col_names)
```

```python
#Generamos un diccionario vacío donde irá la información procesada desde la URL externa
counter = 0
main_dict ={}
for col in col_names:
    main_dict[col]=[]
```

```python
#Procesamos fila a fila la infomación para ir rellenando el diccionario con los datos oco hicimos antes
for line in lines:
    #Nos saltamos la primera linea que es la que tiene la cabecera y ya tenemos procesada
    if(counter > 0):
        #Dividimos cada string por las comas como elemento separador
        values = line.strip().split(",")
        #Añadimos cada valor a su respectiva columna del diccionario
        for i in range(len(col_names)):
            main_dict[col_names[i]].append(values[i])
    counter += 1

print("El data set tiene %d filas y %d columnas"%(counter, n_cols))
```

```python
#Ahora convertimos el diccionario en DataFrame y comprobamos que los datos son correcots
medals_df = pd.DataFrame(main_dict)
medals_df.head()
```
